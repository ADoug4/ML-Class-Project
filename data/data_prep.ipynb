```python
import itertools
import pandas as pd
import pathlib
from tkinter import Tcl

#data_dir = pathlib.Path("..") / ".." / "pkurlak-scenarios" / "scenarios" / "cut-in" # adjust to your fs layout and chosen scenario
data_dir = pathlib.Path("ego-cut-behind")
scenario_files = list(data_dir.glob("**/*.csv"))
scenario_files = Tcl().call('lsort', '-dict', scenario_files) # sort in natural order, natsort.natsorted wasn't working as expected
features = ['time', ' id', ' x', ' h', ' y', ' speed', ' laneId'] # note: weird space in some feature names after parsing

scenario_dict_list = []

for idx, scenario_file in enumerate(scenario_files):
        
        scenario_path = pathlib.Path(scenario_file)
        scenario_df = pd.read_csv(scenario_path, skiprows=1)
        scenario_df = scenario_df[features]
        scenario_dict = {}

        for _, row in scenario_df.iterrows():

                actor = "ego" if row[' id'] == 0 else "npc"
                timestep = int(float(row['time']) * 10)

                scenario_dict[f"{actor}_x_{timestep}"] = row[' x']
                scenario_dict[f"{actor}_y_{timestep}"] = row[' y']
                scenario_dict[f"{actor}_h_{timestep}"] = row[' h']
                scenario_dict[f"{actor}_speed_{timestep}"] = row[' speed']
                scenario_dict[f"{actor}_laneId_{timestep}"] = row[' laneId']

        scenario_dict["is_cut-in"] = 0
        scenario_dict_list.append(scenario_dict)
        print("done", idx)

dataset = pd.DataFrame(scenario_dict_list)
dataset.to_csv("ego-cut-behind_dataset.csv")
```
